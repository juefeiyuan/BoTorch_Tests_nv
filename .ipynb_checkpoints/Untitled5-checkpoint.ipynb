```python
import os
import tempfile
import copy
import numpy          as np
import numpy.random   as npr
import scipy.linalg   as spla
import scipy.stats    as sps
import scipy.optimize as spo
import multiprocessing
import ast
import numpy as np
import numpy.random as npr
# from sklearn.preprocessing import normalize
import numpy as np
import GPy
#%config InlineBackend.figure_format = 'svg'
from pylab import *
import random
import warnings
import time
import sys


import math
import scipy.linalg as spla
import scipy.optimize as spo
import scipy.stats as sps








def target(X):
    x1 = X[0]
    x2 = X[1]
    
    result = (x2- (5.1/(4*math.pi*math.pi))*x1*x1 + (5/math.pi)*x1 - 6)**2 + 10*(1 - (1/(8*math.pi)))*math.cos(x1) + 10
    return result








def EI(x, Xsamples, Ysamples, GP):
    
    x = np.array([x])
    f_star = np.min(Ysamples)

    pos_mu, pos_var = GP.predict(x)
    pos_mu = pos_mu[0,0]
    pos_var = pos_var[0,0]
    pos_sigma = np.sqrt(pos_var)

    W = (f_star - pos_mu)/pos_sigma
    ei = pos_sigma * (W*sps.norm.cdf(W) + sps.norm.pdf(W))
    result = -ei

    return result







def next_sampling_point(EI, Xsamples, Ysamples, GP, bnds, opt_method, maxiter, nRestart =25):

    min_value = 1000
    point_to_sample = None
    
    objective = functools.partial(EI, Xsamples = Xsamples, Ysamples = Ysamples,  GP = GP)
    


    for i in range(nRestart):
        warnings.filterwarnings("ignore")
        bnds_array = np.array(bnds)
        initial_point = npr.random(2)*15 + bnds_array[:, 0]
        function_to_optimize = objective
        warnings.filterwarnings("error")
        try:
            res = spo.minimize(function_to_optimize, initial_point, method = opt_method,\
                                   options={'maxiter':maxiter, 'disp': False}, bounds = bnds)  
            
            if res.fun < min_value:
                min_value = res.fun
                point_to_sample = res.x   
        except:
            continue

            
    point_to_sample = np.array(point_to_sample)
    
    return point_to_sample

```

```python
import functools



nRestart = 1000
opt_method = 'SLSQP'
maxiter = 10000





d = 2
bnds = [np.array([-5., 10.]), np.array([ 0., 15.])]



Xsamples = npr.rand(3, 2)*15 + np.array([-5.,  0.])
Ysamples = np.zeros((Xsamples.shape[0]))

for i in range(Xsamples.shape[0]):
    Ysamples[i] = target(Xsamples[i,:])
    
Ysamples = np.asarray([Ysamples])
Ysamples = Ysamples.T
```

```python
warnings.filterwarnings('ignore')
initial_value_of_l = None

if initial_value_of_l is None:
    lengthscale = 0.3 * np.ones((d))
else:
    lengthscale = initial_value_of_l * np.ones((d))





iterations = 200
valid_count = 0

for i in range(iterations):
    

    warnings.filterwarnings('ignore')
    k = GPy.kern.RBF(input_dim = d, lengthscale = lengthscale, ARD = True)
    GP = GPy.models.GPRegression(Xsamples, Ysamples, k)
    GP.optimize()



    new_x = next_sampling_point(EI, Xsamples, Ysamples, GP, bnds, opt_method, maxiter, nRestart)
    new_y = np.array([target(new_x)])


    Xsamples = np.vstack((Xsamples, new_x))
    Ysamples = np.vstack((Ysamples, new_y))
    print('The number of iteration is ', i)
    print('Current best value is ', np.amin(Ysamples))

```
